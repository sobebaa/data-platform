```python

import pandas as pd
import numpy as np
from faker import Faker
import random

fake = Faker('ru_RU')
Faker.seed(42)
np.random.seed(42)
random.seed(42)

n_rows = 10000


categories = {
    "Социальная защита": ["Пенсии", "Пособия", "Многодетные семьи"],
    "Здравоохранение": ["Доступ к поликлинике", "Скорая помощь", "Медикаменты"],
    "Образование": ["Школы", "Детсады", "ВУЗы"],
    "Жилищный вопрос": ["Аренда", "Ипотека", "Снос жилья"],
    "Труд и занятость": ["Безработица", "Охрана труда", "Трудовые споры"],
    "ЖКХ": ["Отопление", "Водоснабжение", "Электричество"]
}

regions = [
    "г. Астана", "г. Алматы", "Акмолинская", "Карагандинская", "Туркестанская", 
    "Актюбинская", "Алматинская", "Жамбылская", "Восточно-Казахстанская", 
    "Западно-Казахстанская", "Северо-Казахстанская", "Мангистауская"
]

genders = ["М", "Ж"]
statuses = ["На рассмотрении", "Рассмотрено", "Отклонено", "В работе"]
organs = ["Министерство труда", "Акимат города", "Министерство здравоохранения", "МОН РК", "ЦОН"]
submit_types = ["Физическое лицо", "Юридическое лицо"]
submit_methods = ["Веб-сайт", "Мобильное приложение", "Лично", "Почта"]
languages = ["Русский", "Казахский"]
repeat_flags = ["Да", "Нет"]
answer_channels = ["Email", "Портал", "Телефон", "Бумажное письмо"]
urgency_levels = ["Низкая", "Средняя", "Высокая"]
appeal_types = ["Жалоба", "Предложение", "Вопрос", "Благодарность"]
topics = ["Земельный вопрос", "Медицина", "Образование", "Социальные пособия", "Электроснабжение"]

data = []

for i in range(1, n_rows + 1):
    category = random.choice(list(categories.keys()))
    subcategory = random.choice(categories[category])
    region = random.choice(regions)
    age = np.random.randint(18, 90)
    gender = random.choice(genders)
    status = random.choice(statuses)
    submit_type = random.choice(submit_types)
    submit_method = random.choice(submit_methods)
    language = random.choice(languages)
    is_repeat = random.choice(repeat_flags)
    answer_channel = random.choice(answer_channels)
    urgency = random.choice(urgency_levels)
    appeal_type = random.choice(appeal_types)
    topic = random.choice(topics)
    num_children = np.random.poisson(1.5)
    max_child_age = np.random.randint(0, 25) if num_children > 0 else np.nan
    created_at = fake.date_between(start_date='-2y', end_date='today')
    response_at = created_at + pd.Timedelta(days=random.randint(2, 40))
    organ = random.choice(organs)

    row = [
        i, created_at, category, subcategory, region, age, gender, status,
        response_at, num_children, max_child_age, organ,
        submit_type, submit_method, language, is_repeat, answer_channel,
        urgency, appeal_type, topic
    ]
    data.append(row)

columns = [
    "ID", "Дата обращения", "Категория", "Подкатегория", "Регион",
    "Возраст заявителя", "Пол", "Статус", "Дата ответа", "Кол-во детей",
    "Возраст старшего ребенка", "Ответственный орган",
    "Тип заявителя", "Способ подачи", "Язык обращения", "Повторное обращение",
    "Канал получения ответа", "Категория срочности", "Вид обращения", "Тематика"
]

df = pd.DataFrame(data, columns=columns)
df.to_csv("e_otinish_dummy.csv", index=False)
print("Файл e_otinish_dummy.csv успешно создан!")
```
